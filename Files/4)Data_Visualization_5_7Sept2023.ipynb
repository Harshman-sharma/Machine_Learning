

```
Exploratory data analysis is divided into 3 types based on number of features:
1) Univariate analysis
2) Bivariate analysis
3) Multivariate analysis
```




Univariate Analysis
```
Single variable : can be categorical or numerical
Histograms, bar charts, pie charts, boxplots
```



Bivariate Analysis

```
2 variables : can be either categorical or numerical

Numerical vs numerical - scatterplot,
Numerical vs categorical - ANOVA
categorical vs categorical - Chi squared error test
```



Multivariate Analysis


```
Multiple Analysis
- Clustering Analysis
dimensionality reduction technique - Principal Component Analysis,
Multi corresponding analysis
```

# Subplots
- Multiple graphs in a single plot

````python
x = np.linspace(1, 25, 10)
y = x**2
````

````python
plt.subplot(1,2,1)
plt.plot(x, y, '-b*')
plt.subplot(1,2,2)
plt.plot(x1, y1, "y--<", ms=20, mec='r')
````

````python
plt.subplot(2,2,1)
plt.plot(x, y, '-b*')
plt.subplot(2,2,2)
plt.plot(x1, y1, "y--<", ms=20, mec='r')
plt.subplot(2,2,3)
plt.plot(x1, y1, "g-.", ms=20, mec='r')
plt.subplot(2,2,4)
plt.plot(x, y, '-o')
````

````python
plt.subplot(1,4,1)
plt.plot(x, y, '-b*')
plt.subplot(1,4,2)
plt.plot(x1, y1, "y--<", ms=20, mec='r')
plt.subplot(1,4,3)
plt.plot(x, y, '-o')
plt.subplot(1,4,4)
plt.plot(x1, y1, "g-.", mec='r')
````

### 2 graphs in 1 **FIGURE**

````python
fig = plt.figure()

ax1 = fig.add_axes([.1,.1,.8,.8]) #x-coor, y-coor, x-length, y-length
ax2 = fig.add_axes([.2,.5,.4,.3])

ax1.plot(x,y)
ax2.plot(y,x)
````

# Pie Charts

````python
labels = ["python", "c++", "java", "php"]
students = [100, 85, 120, 44]

# plotting using figure
fig = plt.figure()
ax1 = fig.add_axes([0.1,.1,.8,.8])
ax1.pie(students, labels=labels, autopct="%0.2f%%")
ax1.legend()
````

````python
plt.pie(students, labels=labels, autopct="%0.2f%%", colors = ['orange', 'pink', 'red',"blue"])
plt.legend()
````

````python
#labels = ["python", "c++", "java", "php"]
Explode = [.1, 0, 0, 0.5]
plt.pie(students, labels=labels, autopct="%0.2f%%", explode = Explode)
plt.legend()
````
