### ***IMPUTING***


```
Replacing with Mean, median or mode
```



Incase of numerical data

````python
df7
````

````python
df7["Mental_age"].mean()
````

````python
df7["Mental_age"] = df7["Mental_age"].fillna(df7["Mental_age"].mean())
````

````python
df7
````

In case of categorical data

````python
df8 = pd.DataFrame({"Name": ["Harsh", "Krishna", "Aman", "Mehak", "Vaibhav",np.NAN, np.NaN,"Harsh"], "Mental_age" : [35,25,24,13,20,21,11,22], "Branch" : ["BDS", "CSE", "BCI", "CSE", "ECE","CSE", "ECE", "ECE"], "location" : ["Gurgaon", "Jharkhand", "Gunah", "Mathura", "Delhi", "Mathura", "Delhi", np.nan ]})
````

````python
df8
````

````python
df8["Name"].mode() #The output is a list so, to use this value, access index 0 of the list
````

````python
df8["Name"] = df8["Name"].fillna(df8["Name"].mode()[0])
````

````python
df8
````

````python
df8.Name.value_counts()
````
