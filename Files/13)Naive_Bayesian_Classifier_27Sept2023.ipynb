````python
# @title Splitting Dependent and Independent Variable

X = df.iloc[:, 1:4]
X
````

````python
type(X)
````

````python
Y = df.iloc[:, 4]
Y
````

````python
type(Y)
````

````python
# @title Label Encoding

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
````

````python
X["Gender"] = le.fit_transform(X["Gender"])
````

````python
X.head()
````

````python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
````

````python
# applying fit transform on all columns with 1 command
X_Scaled = pd.DataFrame(ms.fit_transform(X), columns = X.columns)
````

````python
X_Scaled
````

````python
#Splitting dataset into train and test

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_Scaled, Y, test_size=0.2, random_state = 0)
````

````python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
````

# MODEL TRAINING

````python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
````

````python
nb.fit(X_train, Y_train)
````

````python
pred = nb.predict(X_test)
````

````python
pred
````

````python
Y_test
````

````python
df
````

````python
nb.predict(ms.transform([[1, 23, 40000]]))
````

````python
# @title Evaluation of model
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
````

````python
# Accuracy Score
accuracy = accuracy_score(Y_test, pred)
accuracy
````

````python
confusion_matrix(Y_test, pred)
````

````python
pd.crosstab(Y_test, pred)
````

````python
(58+16)/(58+16+6+0)
````

````python
print(classification_report(Y_test, pred))
````

````python
# PRECISION
# Of all postitive predictions, how many are actually positive

# Precision = TP / (TP + FP)
````

````python
# RECALL
# Of all real postive cases, how many are actually predicted postive

# Recall = TP / (TP + FN)
````

````python
# F1 Score
# = (2 * precision * recall) / (precison + recall)
````

````python
# @title ROC - AUC CURVE

prob = nb.predict_proba(X_test)[:,1]
````

````python
prob
````

````python
# roc_curve
fpr, tpr, thresholds = roc_curve(Y_test, prob)
````

````python
 plt.plot(fpr, tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
````
